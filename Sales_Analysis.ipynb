```python
sales['month_year'] = sales['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
sales['month_year'].sample(10)
```

```python
sales.head()
```

```python
sales_by_month = sales.groupby('month_year').agg({'sales': 'sum'}).reset_index()
```

```python
plt.figure(figsize=(15,6))
plt.plot(sales_by_month['month_year'],sales_by_month['sales'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```

```python
products_sales = sales.groupby('product_name')['sales'].sum().reset_index()
products_sales = products_sales.sort_values('sales', ascending=False)
```

```python
products_sales[:10]
```

```python
products_by_quantity = sales.groupby('product_name')['quantity'].sum().reset_index()
products_by_quantity_sorted = products_by_quantity.sort_values('quantity', ascending=False)
```

```python
products_by_quantity_sorted[:10]
```

```python
plt.figure(figsize=(10,8.5))

sns.countplot(sales['ship_mode'])
plt.show()
```

```python
ship_mode = sales['ship_mode'].value_counts().reset_index()

ship_mode.columns = ['shipping_mode', 'count']

ship_mode.head()
```

```python
fig = plt.figure(figsize=(10, 7))
plt.pie(
    ship_mode['count'],
    labels=ship_mode['shipping_mode'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Orders by Shipping Mode')
plt.axis('equal')
plt.show()
```
